```python
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('Train.csv')
test=pd.read_csv('Test.csv')
data.head()
```

```python
X=pd.concat([data,test],axis=1)
```

```python
X = data.drop(['ID'], axis = 1)
```

```python
X.head()
```

```python
dat_op_list=X['DATOP']
timespam=pd.to_datetime(dat_op_list)
year=timespam.dt.year
month=timespam.dt.month
day=timespam.dt.day
yeardatop=pd.DataFrame(year)
monthdatop=pd.DataFrame(month)
daydatop=pd.DataFrame(day)
std_list=X['STD']
timespam=pd.to_datetime(std_list)
year=timespam.dt.year
month=timespam.dt.month
day=timespam.dt.day
hour=timespam.dt.hour
minutes=timespam.dt.minute
sec=timespam.dt.second
yearstd=pd.DataFrame(year)
monthstd=pd.DataFrame(month)
hourstd=pd.DataFrame(hour)
minstd=pd.DataFrame(minutes)
secstd=pd.DataFrame(sec)
daystd=pd.DataFrame(day)
sta_list=X['STA']
timespam=pd.to_datetime(std_list)
year=timespam.dt.year
month=timespam.dt.month
day=timespam.dt.day
hour=timespam.dt.hour
minutes=timespam.dt.minute
sec=timespam.dt.second
yearsta=pd.DataFrame(year)
monthsta=pd.DataFrame(month)
hoursta=pd.DataFrame(hour)
minsta=pd.DataFrame(minutes)
secsta=pd.DataFrame(sec)
daysta=pd.DataFrame(day)
```

```python
var=pd.concat([yeardatop,monthdatop,daydatop,yearstd,monthstd,hourstd,minstd,secstd,daystd,yearsta,monthsta,hoursta,minsta,secsta,daysta],axis='columns')
```

```python
var.columns=['yeardatop','monthdatop','daydatop','yearstd','monthstd','hourstd','minstd','secstd','daystd','yearsta','monthsta','hoursta','minsta','secsta','daysta']
```
